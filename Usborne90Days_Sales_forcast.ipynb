```python
import pandas as pd

# Load the data
data = pd.read_csv("/content/Usborne sales data.csv", encoding="ISO-8859-1", dtype={"Sku": str})

# Convert the 'SalesDate' column to a datetime type
data['SalesDate'] = pd.to_datetime(data['SalesDate'])

# Group the data by SKU and then aggregate sales by date for each SKU
sku_grouped = data.groupby(['Sku', 'SalesDate']).agg({'SalesQuantity': 'sum'}).reset_index()

# Create a dictionary of dataframes for each SKU
sku_dataframes = {sku: df.drop(columns='Sku').rename(columns={'SalesDate': 'ds', 'SalesQuantity': 'y'})
                  for sku, df in sku_grouped.groupby('Sku')}
```

```python
from prophet import Prophet

# List to store forecast dataframes for each SKU
forecast_dfs = []

for sku, sku_data in sku_dataframes.items():
    # Check if SKU data has at least two non-NaN rows
    if sku_data['y'].notna().sum() >= 2:
        # Initialize the Prophet model
        model = Prophet()

        # Fit the model with the SKU-specific data
        model.fit(sku_data)

        # Create a dataframe with future dates (3 months in this case)
        future = model.make_future_dataframe(periods=90)

        # Use the model to make predictions
        forecast = model.predict(future)

        # Add the SKU column to the forecast dataframe
        forecast['Sku'] = sku

        # Append the forecast dataframe to the list
        forecast_dfs.append(forecast)
    else:
        print(f"Skipping SKU {sku} due to insufficient data.")
```

```python
# Concatenate all forecast dataframes into a single dataframe
all_forecasts = pd.concat(forecast_dfs, ignore_index=True)

# Export the concatenated dataframe to a CSV file
all_forecasts.to_csv('all_sku_forecasts.csv', index=False)
```

```python
import pandas as pd

# Load the exported forecast data
forecast_data = pd.read_csv("all_sku_forecasts.csv",dtype={"Sku": str})

# Calculate the total rounded forecasted sales for each SKU over the next 90 days
total_sales_per_sku_rounded = forecast_data.groupby('Sku').apply(lambda x: round(x['yhat'].tail(90).sum())).reset_index()
total_sales_per_sku_rounded.columns = ['Sku', 'Rounded Total Forecasted Sales (Next 90 days)']

# Export the results to a CSV file
total_sales_per_sku_rounded.to_csv('total_sales_forecasts.csv', index=False)


```
